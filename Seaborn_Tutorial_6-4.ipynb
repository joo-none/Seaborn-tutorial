# 6-4. Showing multiple relationships with facets

이전의 함수들이 여러 변수들을 한번에 시각화 할 수는 있지만, 항상 효율적인 것은 아님.

그렇다면, 하나 이상의 변수에 의한 두 변수 간의 관계를 이해하기 위해서는 어떻게 해야할까?

가장 좋은 접근은 여러 그래프를 만드는 것이다. 'relplot()'은 FacetGrid를 기반하고 있으므로 쉽게 이를 해결할 수 있다. 추가적인 변수의 영향을 보여주기 위해서는 'Facet'을 이용하면 된다. 

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time",
)
```

두 변수의 영향도 이러한 방법으로 보여줄 수 있다. 

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="subject",
    col="region", row="event", height=3,
    estimator=None
)
```

다양한 level의 변수의 영향을 측정하고 싶다면 column에 있는 변수들을 facet하고, 그 facet들을 row로 wrap하면 된다.

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```
