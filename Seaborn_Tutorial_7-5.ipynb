## Visualizing bivariate distributions

지금까지의 예시들은 univariate distribution(distribution of single variable, perhaps conditional on a second variable assigned to 'hue')으로 간주된다.

두 번째 변수를 y에 할당하는 것은 bivariate distribution 그래프를 그린다. 

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
tips = sns.load_dataset("tips")
diamonds = sns.load_dataset("diamonds")

```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm")
```

Bivariate histogram은 2차원 좌표계에 데이터를 구간 차원으로 나누어 각 사각형에 관측값을 색상으로 표현한다. 이는 히트맵과 유사하다. 마찬가지로 Bivariate KDE 플롯은 2D 좌표계에 관측값을 평활화한다. 

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
```

'hue'를 할당하는 것은 여러 개의 히트맵이나 contour sets을 다양한 색을 사용하여 표현한다. Bivariate 히스토그램에서, 이는 조건부 분포 간에 최소한의 간섭이 있을 때에만 잘 작동한다. 

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

Bivariate KDE의 윤곽 접근 방식은 (너무 많은 경계가 있으면 복잡해지지만) 중첩을 표현하는 데에 더 적합하다.

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

단변량 플롯과 마찬가지로, 구간차원의 크기(bins) 또는 smoothing bandwidth는 그래프가 분포를 얼마나 잘 나타내는지 결정한다. 동일한 매개변수가 적용되지만 한 쌍의 값을 지정하여 각 변수에 맞게 적용할 수 있다. 

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", binwidth=(2, .5))
```

히트맵에 interpretation을 추가하기 위해서는 colorbar를 설정해주면 된다. 

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", binwidth=(2, .5), cbar=True)
```

이변량 밀도 윤곽의 의미는 단순하지 않다. 밀도는 직관적으로 해석하기가 어렵고, 윤곽은 밀도의 등비 비율로 그려지기 때문이다. 이는 각각의 곡선이 밀도의 부분의 단계를 보여준다는 것을 의미한다. 'p' values는 임계값 매개변수에 의해 제어되는 최저 레벨과 레벨에 의해 재어되는 숫자로 인해 균등한 간격을 유지한다. 

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde", thresh=.2, levels=4)
```

'levels' 파라미터는 더 높은 수준의 control을 위해 list 값을 받기도 한다.

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde", levels=[.01, .05, .1, .8])
```

이변수 히스토그램은 두 변수 중 하나 또는 두 변수 모두가 불연속형인 것을 허용한다. 하나의 불연속형 변수와 하나의 연속형 변수를 그리면 조건부 단변량 분포를 비교할 수 있는 또다른 방법이 된다.

```python
sns.displot(diamonds, x="price", y="clarity", log_scale=(True, False))
```

반면, 두 개의 불연속형 변수를 그린 그래프는 관측값의 cross-tabulation을 표현하는 손쉬운 방법이 된다. 

```python
sns.displot(diamonds, x="color", y="clarity")
```
